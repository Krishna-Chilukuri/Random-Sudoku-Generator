```python
from math import sqrt, floor
from random import random


sudokuGrid = [[0 for i in range(9)] for j in range(9)]
numList = [1,2,3,4,5,6,7,8,9]
gridSize = 9
gridRoot = int(sqrt(gridSize))
```

```python
def checkGrid(i, j, num):
    return checkInRow(i, num) and checkInCol(j, num) and checkInBox(i- i % gridRoot, j - j % gridRoot, num)

def checkInRow(row, num):
    for col in range(gridSize):
        if sudokuGrid[row][col] == num:
            return False
    return True

def checkInCol(col, num):
    for row in range(gridSize):
        if sudokuGrid[row][col] == num:
            return False
    return True

def checkInBox(row, col, num):
    for i in range(gridRoot):
        for j in range(gridRoot):
            if sudokuGrid[row + i][col + j] == num:
                return False
    return True
```

```python
def fillElement(row, col):
    num = 0
    for i in range(gridRoot):
        for j in range(gridRoot):
            # shuffle(numList)
            # for num in numList:
            while True:
                num = floor(random() * gridSize + 1)
                if checkInBox(row, col, num):
                    break
            sudokuGrid[row + i][col + j] = num

def generateDiagonalElements():
    for i in range(0, gridSize, gridRoot):
        fillElement(i, i)


def generateNonDiagElements(i, j):
    if i == gridSize - 1 and j == gridSize:
        return True
    if j == gridSize:
        i += 1
        j = 0
    
    if sudokuGrid[i][j] != 0:
        return generateNonDiagElements(i, j + 1)
    for num in numList:
        if checkGrid(i, j, num):
            sudokuGrid[i][j] = num
            if generateNonDiagElements(i, j+1):
                return True
            sudokuGrid[i][j] = 0

def removeElements(removalCount):
    while removalCount > 0:
        randRow = floor(random() * gridSize)
        randCol = floor(random() * gridSize)
        if sudokuGrid[randRow][randCol] != 0:
            sudokuGrid[randRow][randCol] = 0
            removalCount -= 1

def generateSudoku():
    generateDiagonalElements()
    generateNonDiagElements(0, gridRoot)
    # removeElements(removalCount)
```
